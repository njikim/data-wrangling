```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

### 9개의 xlsx를 불러와서 list에 담기

```python
import glob,os

path = r'C:/Python/movie-data-analysis/data/movie' 

all_files = glob.glob(os.path.join(path, "*.csv")) 
print(all_files)
li = []
for filename in all_files:
    df = pd.read_csv(filename, skiprows=[0,1,2,3,5])
    li.append(df)
```

### 각 df마다 상위 100개만 뽑아서 다시 저장

```python
length = range(len(li))
```

```python
for i in length:
    li[i] = li[i][:100]
```

```python
# 상위 백개씩 잘 들어감
li[0].shape
li[2].shape
```

### li에 담긴 각 df 마다 해당 연도 컬럼 추가

```python
for i in length:
    li[i]['연도'] = int(str(201) + str(i))
```

```python
# 잘 들어갔는지 체크
for i in length:
    print(li[i]['연도'].unique())
```

### 이제 하나의 프레임으로 합치기

```python
frame = pd.concat(li, axis=0, ignore_index=True)
frame.head()
```

### 컬럼 정리: space 지우기, 매출액 to 매출액 점유율

```python
frame.columns
```

```python
frame.columns = ['순위', '영화명', '개봉일', '매출액', '매출액 점유율', '누적매출액', '관객수', '누적관객수', '스크린수',
       '상영횟수', '대표국적', '국적', '제작사', '배급사', '등급', '장르', '감독', '배우', '연도']
```

```python
frame.columns
```

## 계층적 색인
1차 인덱스: 연도별 (based on 생성한 연도 컬럼) - 2차 인덱스: 순위별

```python
frame.index=[
             [frame['연도'][i] for i in range(len(frame['연도']))],
             [frame['순위'][i] for i in range(len(frame['순위']))] ]
frame.head()
```

```python
frame.index.names=['year','rank']
frame.head()
```

```python
frame.drop(['순위','연도'], axis=1, inplace=True)
```
