```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

### 9개의 xlsx를 불러와서 list에 담기

```python
import glob,os

path = r'C:/Python/movie-data-analysis/data/movie' 

all_files = glob.glob(os.path.join(path, "*.csv")) 

li = []
for filename in all_files:
    df = pd.read_csv(filename, skiprows=[0,1,2,3,5])
    li.append(df)
```

### 각 df마다 상위 100개만 뽑아서 다시 저장

```python
length = range(len(li))
```

```python
for i in length:
    li[i] = li[i][:100]
```

```python
# 상위 백개씩 잘 들어감
li[0].shape
li[2].shape
```

### li에 담긴 각 df 마다 해당 연도 컬럼 추가

```python
for i in length:
    li[i]['연도'] = str(201) + str(i)
```

```python
# 잘 들어갔는지 체크
for i in length:
    print(li[i]['연도'].unique())
```

### 이제 하나의 프레임으로 합치기

```python
frame = pd.concat(li, axis=0, ignore_index=True)
frame.head()
```

```python
type(frame['연도'][0])
```

# <font color=blue> 필요 컬럼 string에서 , 제거 후 date, int 타입으로 바꾸기 </font>

```python
# 쉼표제거
frame['매출액'] = frame['매출액'].apply(lambda x: x.replace(',', ''))
frame['누적매출액']= frame['누적매출액'].apply(lambda x: x.replace(',', ''))
frame['관객수'] = frame['관객수'].apply(lambda x: x.replace(',', ''))
frame['누적관객수'] = frame['누적관객수'].apply(lambda x: x.replace(',', ''))
frame['스크린수'] = frame['스크린수'].apply(lambda x: x.replace(',', ''))
frame['상영횟수'] = frame['상영횟수'].apply(lambda x: x.replace(',', ''))
```

```python
# int 형식으로 바꿔줄 컬럼
frame[['순위','매출액','누적매출액','관객수','스크린수','상영횟수','연도']] = frame[['순위','매출액','누적매출액','관객수','스크린수','상영횟수','연도']].apply(pd.to_numeric)
```

```python
# 날짜 형식으로 바꿔줄 컬럼
frame['개봉일'] = frame['개봉일'].apply(pd.to_datetime)
```
